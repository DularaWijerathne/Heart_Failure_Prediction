```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split

from sklearn.impute import SimpleImputer

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

from sklearn.ensemble import RandomForestClassifier

from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer

from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

```

## Model Implementation

```python
# define independent and dependent variables

X = df.iloc[:, :11]
y = df['HeartDisease']
```

```python
# devide dataset in to train and test

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=12)
```

```python
# define preprocessor

# define numerical processing pipeline
numeric_processor = Pipeline(steps = [('imputer_mean', SimpleImputer(missing_values=np.nan, strategy='mean')),
                          ('scaler', StandardScaler())])

# define categorical processing pipeline
categoric_processor = Pipeline(steps = [('imputer_constant', SimpleImputer(fill_value='missing', strategy='constant')),
                          ('encoder', OneHotEncoder(drop='first', sparse_output=False))])

# combibe numerical and categorical processors together
preprocessor = ColumnTransformer([('numeric_processor', numeric_processor, ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']), 
                         ('categorical_processor', categoric_processor, ['Sex', 'ChestPainType', 'FastingBS', 'RestingECG', 'ExerciseAngina', 'ST_Slope'])])

preprocessor
```

```python
# define pipeline

pipe = make_pipeline(preprocessor, RandomForestClassifier())
pipe
```

```python
# train model

pipe.fit(X_train, y_train)
```

```python
# make preductions

y_predict = pipe.predict(X_test)
y_predict
```

```python
# evaluate performance

print(f'Classification Report \n{classification_report(y_test, y_predict)}')
print(f'Accuracy {int(accuracy_score(y_test, y_predict)*100)}%')
```

## Model Validation

```python
cross_validations = cross_val_score(pipe, X, y, cv=KFold(10, shuffle=True, random_state=12))
cross_validations
```

```python
print(f'Accuracy of the Model : {int(np.mean(cross_validations)*100)}%')
```
